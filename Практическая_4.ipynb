```
class style_of_presentation: #Класс Стиль изложения
    def __init__(self, s):
        self.style = s  #Стиль изложения

class prose(style_of_presentation): #Класс-наследник "Проза"
    def __init__(self, s, p):
        super().__init__(s)
        self.prose = p  #жанр прозы

class rhyme(style_of_presentation): #Класс-наследник "Рифма"
    def __init__(self, s, r):
        super().__init__(s)
        self.rhyme = r  #вид рифмы
class poem(rhyme): #Класс стихотворение
    def __init__(self, s, r, a, n):
        super().__init__(s, r)
        self.author = a  #автор
        self.name = n #название

    def info(self):
        return "{0}, {1}, {2}, {3}".format(self.author, self.style, self.rhyme, self.name)    
```

```
poem_1 = poem('рифма', 'амфибрахий', 'Пушкин', "Осень")
print(poem_1.info())
```
